```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Configure the fonts 
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "cm" 

fig, ax = plt.subplots(figsize=(12, 8))

# Create margin values and compute the hinge loss
margin = np.linspace(-3, 4, 400)
hinge_loss = np.maximum(0, 1 - margin)

# Configure axes at origin
ax.spines["left"].set_position("zero")
ax.spines["left"].set_linewidth(2.0)
ax.spines["bottom"].set_position("zero")
ax.spines["bottom"].set_linewidth(2.0)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# Set limits and labels
ax.set_xlim(-3, 4)
ax.set_ylim(0, 4.1)
ax.set_xlabel(r"$y_i(w^Tx_i+b)$", fontsize=24)                  
ax.set_ylabel(r"$\ell(x_i,y_i;w,b)$", fontsize=24, rotation=90) 
ax.xaxis.set_label_coords(0.91, -0.05)
ax.yaxis.set_label_coords(0.405, 0.85)

# Configure ticks
ax.set_xticks([-3, -2, -1, 0, 1, 2, 3, 4])
ax.set_yticks([1, 2, 3, 4])
ax.tick_params(labelsize=16, direction="in", length=8, width=1.2)

# Plot the hinge loss
ax.axvline(x=1, color="k", linestyle="--", linewidth=2.0, alpha=0.7)
ax.plot(margin, hinge_loss, color="#E63946", linewidth=3.5, zorder=10, clip_on=False)

# Add region labels and shading
ax.text(-1.4, 3.8, "Loss Penalty Region", fontsize=20, ha="center", color="#C41E2A")
ax.text(2.5, 3.8, "No Penalty Region", fontsize=20, ha="center", color="#2C5A7A")
ax.axvspan(-3, 1, alpha=0.15, color="#E63946") # Loss Penalty Region
ax.axvspan(1, 4, alpha=0.15, color="#06D6A0")  # No Penalty Region

plt.tight_layout()

# plt.savefig(
#     "hinge_loss.svg",
#     format="svg",
#     bbox_inches="tight",
#     dpi=300,
#     transparent=False,
# )

plt.show()
```
